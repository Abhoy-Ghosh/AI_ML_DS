```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# feature engineering

df['date'] = df['Date_of_Journey'].str.split("/").str[0].astype(int)

df['month'] = df['Date_of_Journey'].str.split("/").str[1]

df['year'] = df['Date_of_Journey'].str.split("/").str[2]
```

```python
df['month'] = df.month.astype(int)

df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
# drop the date of journey column

df.drop("Date_of_Journey", axis =1, inplace = True)
```

```python
df.info()
```

```python
df["Arrival_Time"] = df.Arrival_Time.apply(lambda x : x.split(' ')[0])
```

```python
df['arrival_hour'] = df['Arrival_Time'].str.split(":").str[0].astype(int)
df['arrival_min'] = df['Arrival_Time'].apply(lambda x:x.split(" ")[0]).str.split(":").str[1].astype(int)

df.head(2)

```

```python
df.drop('Arrival_Time',axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
df['dep_hour'] = df['Dep_Time'].str.split(":").str[0].astype(int)
df['dep_min'] = df['Dep_Time'].str.split(":").str[1].astype(int)


df.head(2)
```

```python
df.drop('Dep_Time',axis = 1, inplace = True)
```

```python
df.head(3)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['duration_hour'] = df['Duration'].str.split(" ").str[0].str.split("h").str[0]
df['duration_min'] = df['Duration'].str.split(" ").str[1].str.split("m").str[0]

```

```python
df.head()
```

```python
df['duration_min'] = df['duration_min'].fillna(0).astype(int)
# df['duration_hour'] = df['duration_hour'].fillna(0).astype(int)

#df['duration_min'] = df['duration_min'].apply(lambda x: 1 if pd.isna(x) else x)
#df['duration_min'] = df['duration_min'].replace(np.nan, 1)
```

```python
df[df['duration_min'].isna()]
df[df['duration_hour'].isna()]
```

```python
df['duration_hour'] = df['Duration'].str.extract(r'(\d+)h')[0].fillna(0).astype(int)
# df['duration_min']  = df['Duration'].str.extract(r'(\d+)m')[0].fillna(0).astype(int)
```

```python
df.info()
```

```python
df['total_duration']= df['duration_hour']*60 + df['duration_min']
```
