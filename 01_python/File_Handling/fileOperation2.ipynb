```python
# read a whole file

with open("example.txt","r") as file:
    content = file.read()
    print(content)
```

```python
# write  a file (over writing)

with open("ex.txt","w") as file:
    content = file.write("i am a python lover")
    print(content)
# till now it create a new file name ex.txt
# and in content number of letter and spaces is stored


with open("ex.txt","w") as file:
    file.write("i am a python lover")
    file.write("i am a python lover")
# here  over writing multiple lines in the file 
```

```python
# read a whole file but line by line

with open("example.txt","r") as file:
    for line in file:
    # content = file.read()
        print((line).strip()) # .strip() remove the extra file 
```

```python
# writing a list of lines to a file

lines =  ["\nfirst line\n","secondline "]

with open("ex.txt","a") as file:
    file.writelines(lines)
```
