```python
#check the file exists or not

path = "exampleq.txt"

if os.path.exists(path):
    print(f"file exist {path}")

else:
    print(f"do not exist")
```

```python
#check the file exists or not also dir or file

path = "exampleq.txt"

if os.path.isdir(path):
    print(f"directory exist {path}")
elif os.path.isfile(path):
    print(f"file exist {path}")
else:
    print(f"do not a file or a folder")
```
