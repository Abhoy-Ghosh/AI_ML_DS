# File Operation In Python

## Introduction to File Operations in Python
In this lecture, we continue our discussion on Python, focusing on file operations. We will perform various file operations using Python, working with both text and binary files. We will learn how to create files, write to them, read their content, and more.

## Reading a Whole File
To read a whole file, we use the with open method. We specify the file name, such as example.txt, and use the 'r' parameter, which stands for read mode. Other parameters like write, append, and read-append are also available, but for now, we focus on read mode.

```python Code Sample
with open('example.txt', 'r') as file:
    content = file.read()
```
If the file does not exist, Python will raise an error indicating that there is no such file or directory. To avoid this, ensure the file exists before attempting to read it.

## Creating and Writing to a File
To create a file and write content to it, use the write mode. For example, to create example.txt and write some lines:

``` python Code Sample
with open('example.txt', 'w') as file:
    file.write('Hello.\n')
    file.write('How are you.\n')
    file.write('I am good.\n')
    file.write('Krish is my name.\n')
    file.write('Welcome to the course.\n')
```
## Reading a File Line by Line
To read a file line by line, use a for loop with the file object. This allows you to process each line individually.

```python Code Sample
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```
When reading line by line, new line characters are included. To remove them, use the .strip() method.

```python Code Sample
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```
## Writing to a File and Overwriting
To write to a file and overwrite its content, open the file in write mode ('w'). Any existing content will be erased.

```python Code Sample
with open('example.txt', 'w') as file:
    file.write('Hello World\n')
    file.write('This is a new line\n')
```
Opening a file in write mode will overwrite all existing content. To avoid overwriting, use append mode.

## Appending to a File
To add content to the end of a file without overwriting, use append mode ('a').

```python Code Sample
with open('example.txt', 'a') as file:
    file.write('Append operation taking place\n')
```
Each time you append, you can add a new line character (\n) to ensure the text appears on a new line.

## Writing Multiple Lines to a File
To write a list of lines to a file, use the writelines() method.

``` python Code Sample
lines = ['This is my first line\n', 'Second line\n', 'Third line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
Working with Binary Files
```
Binary files, such as those with a .bin extension, are handled using binary modes. To write to a binary file, use 'wb' mode. To read from a binary file, use 'rb' mode.

```python Code Sample
with open('example.bin', 'wb') as file:
    file.write(b'Hello world')
python Code Sample
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

## Copying Content from One File to Another
To copy content from a source text file to a destination text file, read from the source and write to the destination.

```python Code Sample
with open('example.txt', 'r') as source_file:
    content = source_file.read()
with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```
Counting Lines, Words, and Characters in a File
To count the number of lines, words, and characters in a text file, read the lines and process them accordingly.

```python Code Sample
def count_file_stats(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        num_words = sum(len(line.split()) for line in lines)
        num_chars = sum(len(line) for line in lines)
    return num_lines, num_words, num_chars
```

Reading and Writing in the Same File Session
To open a file for both reading and writing, use the 'w+' mode. If the file does not exist, it will be created. If it exists, its content will be truncated (overwritten).

``` python Code Sample
with open('example.txt', 'w+') as file:
    file.write('Hello world\n')
    file.write('This is a new line\n')
    file.seek(0)
    content = file.read()
    print(content)
```
The seek(0) method moves the file cursor to the beginning, allowing you to read the content after writing. Always use seek(0) before reading if you have just written to the file.
