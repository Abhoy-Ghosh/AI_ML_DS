```python
#  data serialization

# to convert dictionary or any  to JSON
# Api give response in JSON format we need to manipulate them

import json

data = {
    "name" : "python",
    "work" : "AI"
}

json_str = json.dumps(data) # dict -> JSON //JSON ARE STRING IN TYPE
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str) # JSON -> dict
print(parsed_data)
print(type(parsed_data))
```
