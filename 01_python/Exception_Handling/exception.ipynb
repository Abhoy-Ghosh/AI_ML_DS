## Exception Handling in Python

**What is it?**  
Exception handling is a way to gracefully handle errors during program execution so the program doesn’t crash.  
It uses `try`, `except`, `else`, and `finally` blocks.

**Key Blocks:**
- `try` → code that might raise an error  
- `except` → handles the error  
- `else` → runs if no error occurs  
- `finally` → always runs (cleanup code, e.g., closing files)

---

### Common Scenarios & Exceptions

| Exception              | When it Occurs                        |
|------------------------|----------------------------------------|
| `ZeroDivisionError`    | Dividing a number by zero             |
| `FileNotFoundError`    | File/Directory doesn’t exist           |
| `ValueError`           | Invalid type conversion (e.g., `int("abc")`) |
| `IndexError`           | Accessing an invalid list index        |
| `KeyError`             | Accessing a missing dictionary key     |
| `TypeError`            | Operation on incompatible data types   |

---

**Example:**
```python
try:
    num = int("abc")   # error
except ValueError:
    print("Invalid number!")
finally:
    print("Done")

````python
try:
    num = int(input("enter a num : "))
    res = 10 /num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("please give a valid number in dominator")
except Exception as ex:
    print(ex)
````

````python
# try,except and else

try:
    num = int(input("enter a num : "))
    res = 10 /num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("please give a valid number in dominator")
except Exception as ex:
    print(ex)
else:  # it execute when there is no exception
    print(f"the result is {res}")
````

````python
# try , except, else and finally

try:
    num = int(input("enter a num : "))
    res = 10 /num
except ValueError:
    print("this is not a valid number")
except ZeroDivisionError:
    print("please give a valid number in dominator")
except Exception as ex:
    print(ex)
else:  
    print(f"the result is {res}")
finally: # it executes wheather there is a exception or not
    print("execution completed")
````

````python
# file handling and exception handling 

try:
    file = open ("example.txt","r")
    content = file.read()
    print(content)

except FileNotFoundError as ex1:
    print(ex1)
    print("the file is not found")

finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print("file is closed")
        
    
````
