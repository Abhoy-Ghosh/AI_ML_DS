# Classes And Objects In Python

## Introduction to Classes and Objects in Python

In this session, we continue our discussion on Python, focusing on the
object-oriented programming (OOP) module. We will discuss classes and
objects, which are fundamental to modeling real-world scenarios in
Python applications.

## Object-Oriented Programming Paradigm

Object-oriented programming is a paradigm that uses objects to design
applications and computer programs. OOP allows us to model real-world
scenarios using classes and objects. For any end-to-end project, it is
essential to use classes and objects to increase code reusability and
efficiency.

## Defining a Class in Python

A class is a blueprint for creating objects. It contains attributes and
methods. For example, to define a class named Car, we use the class
keyword followed by the class name.

``` python
class Car:
    pass
```

## Creating Objects from a Class

Objects are instances of a class. For example, we can create objects
such as Audi and BMW from the Car class. These objects are created using
the class as a blueprint.

``` python
Audi = Car()
BMW = Car()
```

## Understanding Object Types

If we print the type of Audi, it will show that it is an object of the
Car class. This confirms that Audi and BMW are objects instantiated from
the Car class.

``` python
print(type(Audi))
print(type(BMW))
```

## Attributes and Instance Variables

Attributes, also known as instance variables, are properties associated
with objects. For example, we can assign the number of windows to the
Audi object.

``` python
Audi.windows = 4
print(Audi.windows)
```

## Proper Way to Define Attributes

Defining attributes directly on objects is not a proper way, as
different objects may have different attributes. For example, if we
create a Tata object and assign it a doors attribute, it will not have
the windows attribute unless explicitly defined.

``` python
Tata = Car()
Tata.doors = 4
print(Tata.windows)  # This will raise an AttributeError
```

## Using the Constructor (**init**) to Define Attributes

A better way to define attributes is by using the constructor method
**init**. This ensures all objects have the same set of attributes
initialized when they are created.

``` python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

## Creating Objects with Attributes

When creating a Dog object, we must provide the name and age attributes.
The self keyword is used to refer to the instance variables within the
class.

``` python
dog1 = Dog('Buddy', 3)
print(dog1.name)
print(dog1.age)

dog2 = Dog('Lucy', 4)
print(dog2.name)
print(dog2.age)
```

## Instance Methods

Instance methods are functions defined within a class that operate on
the instance variables. For example, a dog can have a method to bark.

``` python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog('Buddy', 3)
dog1.bark()
dog2 = Dog('Lucy', 4)
dog2.bark()
```

## Modeling a Bank Account Example

Let us design a more practical example: modeling a bank account. In a
bank account, you can deposit money, withdraw money, and check the
balance. We will use a class to represent a bank account with these
functionalities.

``` python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} is deposited, new balance is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} is withdrawn, new balance is {self.balance}")

    def get_balance(self):
        return self.balance

account = BankAccount('Krish', 5000)
print(account.balance)
account.deposit(100)
account.withdraw(300)
print(account.get_balance())
```

## Conclusion

Object-oriented programming is essential for modeling real-world
scenarios using classes and objects. In this lesson, we learned how to
create classes and objects, define instance variables and methods, and
use them to perform various operations. Always pay attention to
indentation and proper structuring when defining classes, constructors,
and methods. The self keyword is crucial for accessing instance
variables and methods within the class.

## Key Takeaways

-   Classes in Python serve as blueprints for creating objects,
    encapsulating attributes and methods.
-   Instance variables and methods are defined within classes and
    accessed via objects.
-   The **init** method acts as a constructor to initialize instance
    variables.
-   Using the self keyword allows access to instance variables and
    methods within the class.
-   Proper structuring of attributes and methods enhances code
    reusability and models real-world scenarios effectively.
