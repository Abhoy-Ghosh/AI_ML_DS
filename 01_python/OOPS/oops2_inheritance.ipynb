````python
## inheritance 

## parent class
class Car:
    def __init__(self,doors,windows,engine):
        self.doors =  doors
        self.windows = windows
        self.engine = engine

    def driving(self):
        print(f"the car has {self.doors} doors, {self.windows} windows, {self.engine} engine type")


lamborgini = Car(2,4,"petrol")
lamborgini.driving()
        
````

````python
# Single inheritance
#new child class
class Tesla(Car):
    def __init__(self, doors, windows, engine,is_selfdriving):
        super().__init__(doors, windows, engine)
        self.is_selfdriving = is_selfdriving
    
    def features(self):
        return f"supports self driving: {self.is_selfdriving}"
    
tesla_truck = Tesla(4,4,"electric",True)

# print(tesla_truck.features) => <bound method Tesla.features of <__main__.Tesla object at 0x000001E8F6E123C0>>
print(tesla_truck.features())

# print(tesla_truck.driving) => <bound method Car.driving of <__main__.Tesla object at 0x000001E8F6E123C0>>
print(tesla_truck.driving())
````

````python
# multiple inheritance 

# base class 1 = Animal
class Animal:
    def __init__(self,name):
        self.name = name

    def say_name(self):
        return f'{self.name} is name of the animal'
    

# base class 2 = Pet
class Pet:
    def __init__(self,owner):
        self.owner =  owner

    def say_owner(self):
        return f"{self.owner} is owner of the pet"
````

````python
# derived class

class Cat(Animal,Pet):
      
    def __init__(self, name, owner, breed):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
        self.breed = breed
    
    def say_details(self):
        return f"it is {self.breed} named {self.name} owned by {self.owner}"
````

````python
cat1 = Cat( "laddu", "abc", "cat ")
print(cat1.say_details())

print(cat1.say_name())
````

# Inheritance In OOPS

## Introduction to Inheritance in Python

Inheritance is a fundamental concept in object-oriented programming that
allows a class to inherit attributes and methods from another class. In
this article, we will discuss how to implement inheritance in Python,
covering both single and multiple inheritance with examples.

Inheritance is derived from real-world scenarios. For example, you
inherit property from your parents, such as a house or land, but the
reverse is not possible. Similarly, in Python, inheritance works in the
same way, where a child class inherits properties from a parent class.

## Defining a Parent Class

Let's define a parent class called Car. This class will have attributes
such as windows, doors, and engine type, and an instance method called
drive.

``` python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive the {self.engine_type} car.")
```

We can create an object of the Car class and call its drive method as
follows:

``` python
car_one = Car(4, 5, "petrol")
car_one.drive()
```

**Output:**

    The person will drive the petrol car.

## Single Inheritance Example

Now, let's create a child class Tesla that inherits from the Car class.
Tesla cars have all the attributes of a car, plus an additional
attribute is_self_driving.

``` python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_driving):
        super().__init__(windows, doors, engine_type)
        self.is_self_driving = is_self_driving

    def self_driving_feature(self):
        print(f"Tesla supports self driving: {self.is_self_driving}")
```

We can create an object of the Tesla class and use its methods:

``` python
tesla_one = Tesla(4, 5, "electric", True)
tesla_one.self_driving_feature()
tesla_one.drive()
```

**Output:**

    Tesla supports self driving: True
    The person will drive the electric car.

## Multiple Inheritance

Multiple inheritance occurs when a class inherits from more than one
base class. Let's explore this with an example involving animals and
pets.

``` python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Subclasses must implement this method.")
```

``` python
class Pet:
    def __init__(self, owner):
        self.owner = owner
```

Now, let's create a derived class Dog that inherits from both Animal and
Pet classes.

``` python
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} says woof"
```

We can create an object of the Dog class and test its methods and
attributes:

``` python
dog = Dog("Buddy", "trish")
print(dog.speak())
print(f"Owner: {dog.owner}")
```

**Output:**

    Buddy says woof
    Owner: Crush

## Summary

-   Single inheritance allows a child class to inherit from one parent
    class.
-   Multiple inheritance allows a child class to inherit from multiple
    parent classes.
-   The super() function is useful for calling the parent class
    constructor in single inheritance.
-   In multiple inheritance, parent class constructors can be called
    explicitly using the class name.
-   In the next video, we will model a complete library system using
    single and multiple inheritance.

## Key Takeaways

-   Inheritance in Python allows a class to inherit attributes and
    methods from another class, facilitating code reuse.
-   Single inheritance involves one child class inheriting from one
    parent class, demonstrated with a Car and Tesla example.
-   The super() function is used in child classes to call the parent
    class's constructor and inherit its attributes.
-   Multiple inheritance allows a class to inherit from more than one
    base class, as shown with Animal, Pet, and Dog classes.
