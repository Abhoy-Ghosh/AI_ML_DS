# Encapsulation in OOPS (Python)

## Introduction to Encapsulation
In this video, the discussion continues with respect to Python, focusing on encapsulation and abstraction. This specific session is dedicated to encapsulation, which is an important concept in object-oriented programming. Encapsulation helps design maintainable and reusable code by bundling data and methods that operate on the data within a single unit.

---

## Encapsulation and Its Importance
Encapsulation involves wrapping data (variables) and methods together as a single unit. It restricts direct access to some of the object components, preventing accidental interference and misuse of data. This means that certain variables or methods should not be accessible from outside the class, especially if they are critical to the object's integrity.

---

## Access Modifiers: Public, Protected, and Private

### Public Variables
No underscore prefix. These variables can be accessed from anywhere.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Chris", 34)
print(person.name)  # Output: Chris
```

### Private Variables
To create private variables, use double underscores. These variables cannot be accessed directly from outside the class.

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name
        self.__age = age
        self.gender = gender  # public variable

person = Person("Krish", 34, "male")
print(person.gender)  # Output: male
# print(person.__name)  # This will fail
```

### How Python Handles Private Variables
Private variables are name-mangled in Python. For example, `__age` becomes `_Person__age` internally.

```python
print(person._Person__age)  # Output: 34
```

### Protected Variables
Protected variables use a single underscore. They can be accessed in derived classes.

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name  # protected variable
        self._age = age    # protected variable
        self.gender = gender

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)

employee = Employee("Chris", 34, "male")
print(employee._name)  # Output: Chris
```

## Encapsulation with Getter and Setter Methods
Encapsulation is often implemented using getter and setter methods to access and modify private variables safely.

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # private variable
        self.__age = age    # private variable

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")
```

### Using Getter and Setter Methods
```python
person = Person("Krish", 34)
print(person.get_name())  # Output: Krish
print(person.get_age())   # Output: 34
person.set_age(35)
print(person.get_age())   # Output: 35
person.set_age(-5)        # Output: Age cannot be negative
```

---

## Summary
Encapsulation in Python is achieved by using access modifiers and getter/setter methods. Public variables can be accessed from anywhere, protected variables can be accessed in derived classes, and private variables are restricted to the class itself. Getter and setter methods provide controlled access to private variables, ensuring data integrity.

---

## Key Takeaways
- Encapsulation in Python involves bundling data and methods as a single unit and restricting direct access to some object components.
- Public, protected, and private variables are defined using naming conventions: no underscore for public, single underscore for protected, and double underscore for private.
- Getter and setter methods are used to access and modify private variables safely.
- Private variables cannot be accessed outside the class, while protected variables can be accessed in derived classes.
