# Polymorphism In OOPS

## Introduction to Polymorphism

Polymorphism is a core concept in object-oriented programming that
allows objects of different classes to be treated as objects of a common
superclass. It provides a way to perform a single action in different
forms. The term "polymorphism" literally means "multiple forms."
Typically, polymorphism is achieved through method overriding and
interfaces.

## Method Overriding

Method overriding allows a child class to provide a specific
implementation of a method that is already defined in the parent class.
For example, consider a parent class Animal with a method speak. A
derived class Dog can override this method to provide its own
implementation, such as returning "Woof, woof." Similarly, a Cat class
can override the same method to return "Meow." This demonstrates how the
same method name can have different behaviors in different classes.

``` python
class Animal:
    def speak(self):
        return "The sound of the animal"

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

dog = Dog()
print(dog.speak())  # Output: Woof

cat = Cat()
print(cat.speak())  # Output: Meow
```

## Demonstrating Polymorphism with a Function

We can create a function that accepts an object of the base class and
calls the overridden method. This function can process objects of
different derived classes uniformly, demonstrating polymorphism.

``` python
def animal_speak(animal):
    print(animal.speak())

animal_speak(dog)  # Output: Woof
animal_speak(cat)  # Output: Meow
```

## Polymorphism with Functions and Methods

Polymorphism can also be implemented using functions and methods to
process objects of different types in the same way. For example,
consider a base class Shape with a method area. Derived classes like
Rectangle and Circle override this method to compute their respective
areas.

``` python
class Shape:
    def area(self):
        print("Area of the shape")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

def print_area(shape):
    print(f"The area is {shape.area()}")

rectangle = Rectangle(4, 5)
circle = Circle(3)

print_area(rectangle)  # Output: The area is 20
print_area(circle)     # Output: The area is 28.26
```

## Polymorphism with Abstract Base Classes

In Python, abstract base classes (ABCs) are used to define common
methods for a group of related objects. They enforce that derived
classes implement particular methods, promoting consistency across
different implementations. This supports polymorphism by ensuring that
all subclasses have certain methods.

``` python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"

class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"

car = Car()
motorcycle = Motorcycle()

def start_vehicle(vehicle):
    print(vehicle.start_engine())

start_vehicle(car)        # Output: Car engine started
start_vehicle(motorcycle) # Output: Motorcycle engine started
```

## Conclusion

Polymorphism is a powerful feature that allows flexibility and
integration in code design. It enables a single function to handle
objects of different classes, each with its own implementation of a
method. By understanding and applying polymorphism, you can create
extensible and maintainable object-oriented programs.

# Polymorphism And Abstract Base Classes(interfaces)

Abstarct Base Classes (ABCs) are used to define  common methods for a group of related object. 
They can enforce that Derived Classes implement particular methods ,promoting consistency across different implementations

````python
from abc import ABC,abstractmethod # abstract base classes are imported

class Flower(ABC):
    @abstractmethod
    def smell(self):
        pass

# derived class1
class Rose(Flower):
    def smell(self):
        return "sweet and strong"
    
# derived class2
class Sunflower(Flower):
    def smell(self):
        return "beautiful and but nice"

    
rose = Rose()
print(rose)
print(rose.smell())

# make a function that performs polymorphism
def smell_of_flower(flower_name): #this function takes the object name 
    smell = flower_name.smell()
    print(smell)
    
smell_of_flower(Sunflower())
````
