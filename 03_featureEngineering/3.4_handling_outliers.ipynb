# 5 number summary and BOX plot

###  Min, Max, Median, Q1, Q3, IQR

```python
import numpy as np
```

```python
marks = [14,15,85,52,54,99,45,27,26,74,62,37]

min ,q1, median, q3, max =np.quantile(marks,[0.0,.25,.5,.75,1])
```

```python
min,q1,median,q3,max 
```

```python
IQR =q3-q1
print(IQR)
```

```python
lower_fence = q1 - 1.5*(IQR)
print(lower_fence)

higher_fence = q3 + 1.5*(IQR)
print(higher_fence)
```

```python
import seaborn as sns

sns.boxplot(marks)
```
