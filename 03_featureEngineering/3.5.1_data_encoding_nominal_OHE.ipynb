# 1. Nominal or ONE-Hot encoding
 

 ### sparse matrix created due to  OHE -> overfitiing (Disadvantage)

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame(
    {
        'color' : ["red",'green','red',' blue','yellow','orange']
    }
)
```

```python
df.head()
```

```python
enc = OneHotEncoder()
```

```python
#perform fit and transform

encoded = enc.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
import pandas as pd 
encoder_df  = pd.DataFrame(encoded, columns = enc.get_feature_names_out())
```

```python
encoder_df
```

```python
enc.transform([['red']]).toarray()
```

```python
pd.concat([df,encoder_df],axis= 1)
```
