# Target Guided Ordinal Encoding

```python
df2 =  pd.DataFrame(
    {
        'city' : ['a', 'b', 'c', 'a', 'c', 'b', 'c'],
        'price' : [3,4,2,5,7,1,1]
    }
)

len(df2[['city']])
```

```python
df2
```

```python
mean_price = df2.groupby('city')['price'].mean().to_dict()
```

```python
df2['city_encoded'] = df2['city'].map(mean_price)
```

```python
df2[['city',"city_encoded","price"]]
```

```python
median_price = df2.groupby('city')['price'].median().to_dict()
```

```python
df2['city_encoded_median'] = df2['city'].map(median_price)
```
