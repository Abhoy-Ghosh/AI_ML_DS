What is SMOTE and why use it?

SMOTE, which stands for Synthetic Minority Over-sampling Technique, is a technique used in machine learning to address the issue of imbalanced datasets. Here's a breakdown of SMOTE and its importance:

Imbalanced Datasets: In many datasets, one class (the majority class) significantly outnumbers another class (the minority class). This imbalance can lead machine learning models to be biased towards the majority class, resulting in poor predictive performance for the minority class.

Purpose of SMOTE: SMOTE aims to balance the dataset by generating synthetic instances of the minority class. Instead of simply duplicating existing minority class examples, SMOTE creates new examples by interpolating between existing ones. This approach helps enhance the model's ability to learn features of the minority class.

How SMOTE Works:

SMOTE identifies the nearest neighbors of the minority class instances.
It generates new synthetic samples by selecting points along the line segments connecting a minority class instance with its nearest neighbors.
This leads to the creation of synthetic instances that provide added variance and diversity, improving the model's training process.
Advantages of Using SMOTE:

It helps improve model performance by addressing the bias introduced by imbalanced data.
The synthetic data created through SMOTE can help the model learn better decision boundaries, leading to improved accuracy for the minority class.
Using SMOTE effectively can enhance the performance of machine learning models, especially in scenarios where the minority class is critical for the application's success

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples= 1000, 
                    n_redundant= 0, 
                    n_features=2,
                    n_clusters_per_class= 1,
                    weights=[0.90],
                    random_state= 12)
```

```python
import pandas as pd 

df1 = pd.DataFrame(X,columns=['feature1','feature2'])
df2 = pd.DataFrame(y,columns = ['target'])

final_df = pd.concat([df1,df2],axis =1)
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature1'],final_df['feature2'],c=final_df['target']) 
```

```python
from imblearn.over_sampling import SMOTE
```

```python
## transform the dataset

oversample = SMOTE()

X,y = oversample.fit_resample(final_df[['feature1','feature2']],final_df['target'])
```
